### Comparación entre colegios bilingues y colegios no bilingues

```python
df['cole_bilingue_desc'] = df['cole_bilingue'].map({'S': 'Colegio Bilingüe', 'N': 'Colegio NO Bilingüe'})
bilingual = df[df['cole_bilingue'] == 'S']
non_bilingual = df[df['cole_bilingue'] == 'N']
```

```python
palette = sns.color_palette("Set2")

# Box plots comparativos
plt.figure(figsize=(15, 8))
for i, col in enumerate(['punt_matematicas', 'punt_sociales_ciudadanas', 'punt_c_naturales', 'punt_lectura_critica', 'punt_global']):
    plt.subplot(2, 3, i+1)
    sns.boxplot(x='cole_bilingue_desc', y=col, data=df, palette=palette)
    plt.title(f'Comparación de {col}')
plt.tight_layout()
plt.show()

# Violin plots comparativos
plt.figure(figsize=(15, 8))
for i, col in enumerate(['punt_matematicas', 'punt_sociales_ciudadanas', 'punt_c_naturales', 'punt_lectura_critica', 'punt_global']):
    plt.subplot(2, 3, i+1)
    sns.violinplot(x='cole_bilingue_desc', y=col, data=df, palette=palette)
    plt.title(f'Comparación de {col}')
plt.tight_layout()
plt.show()
```

Prueba de diferencia de medias

```python
means = df.groupby('cole_bilingue_desc')[['punt_matematicas', 'punt_sociales_ciudadanas', 'punt_c_naturales', 'punt_lectura_critica', 'punt_global']].mean()
means.plot(kind='bar', figsize=(10, 6), color=palette)
plt.title('Comparación de medias de puntajes')
plt.ylabel('Puntaje promedio')
plt.xticks(rotation=0)
plt.legend(title='Puntajes')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import ttest_ind
####--------------
#Prueba t para las diferencias de medias
# HO: u1 = u2 --> u1 - u2 = 0
# H1: u1 != u2 --> u1 - u2 != 0
####--------------
for col in ['punt_matematicas', 'punt_sociales_ciudadanas', 'punt_c_naturales', 'punt_lectura_critica', 'punt_global']:
    t_stat, p_val = ttest_ind(bilingual[col], non_bilingual[col], equal_var=False)
    print(f'Test t para {col}: estadístico t = {t_stat:.2f}, p-value = {p_val:.4f}')
```
