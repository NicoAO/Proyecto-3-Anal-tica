### Análisis de los aspectos socioeconómicos

Entonces, las personas cuyos inmuebles pertenezcan a los estratos 1, 2 o 3 significaría que se ubican en la clasificación más baja y en los estratos 5 y 6 se localizan las personas con mayores ingresos económicos (Restrepo, 2024)

```python
estu_fam_bajo = data[
    (data["fami_estratovivienda"].isin([1, 2, 3])) & 
    (data["fami_tienecomputador"] == 0) & 
    (data["fami_tieneinternet"] == 0)
]
estu_fam_medio = data[
    (data["fami_estratovivienda"] == 4) & 
    (data["fami_tienecomputador"] == 1) & 
    (data["fami_tieneinternet"] == 1)
]
estu_fam_alto = data[
    (data["fami_estratovivienda"].isin([5,6])) & 
    (data["fami_tienecomputador"] == 1) & 
    (data["fami_tieneinternet"] == 1)
]
```

```python
plt.figure(figsize=(14, 8))
sns.histplot(estu_fam_bajo['punt_global'], color='blue', label='Bajo', kde=True, stat="density")
sns.histplot(estu_fam_medio['punt_global'], color='green', label='Medio', kde=True, stat="density")
sns.histplot(estu_fam_alto['punt_global'], color='red', label='Alto', kde=True, stat="density")
plt.title('Distribución de Puntaje Global del ICFES por nivel socioeconómico')
plt.xlabel('Puntaje Global del ICFES')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
dataframes = [estu_fam_bajo.assign(Nivel='Bajo'), estu_fam_medio.assign(Nivel='Medio'), estu_fam_alto.assign(Nivel='Alto')]
combined_df = pd.concat(dataframes)
palette = {'Bajo': 'blue', 'Medio': 'green', 'Alto': 'red'}
sns.boxplot(x='Nivel', y='punt_global', data=combined_df, palette= palette)
plt.title('Boxplot de Puntaje Global del ICFES por nivel socioeconómico')
plt.xlabel('Nivel Socioeconómico')
plt.ylabel('Puntaje Global')
plt.grid(True)
plt.show()
```

```python
palette = sns.color_palette("Set2")
plt.figure(figsize=(12, 6))
sns.countplot(x='cole_naturaleza', hue='Nivel', data=combined_df,palette=palette)
plt.title('Distribución de la naturaleza del colegio por nivel socioeconómico')
plt.xlabel('Naturaleza del colegio   0: No oficial  1: Oficial')
plt.ylabel('Cuenta')
plt.legend(title='Nivel Socioeconómico')
plt.grid(True)
plt.show()
```

### Comparación entre colegios bilingues y colegios no bilingues

```python
df['cole_bilingue_desc'] = df['cole_bilingue'].map({'S': 'Colegio Bilingüe', 'N': 'Colegio NO Bilingüe'})
bilingual = df[df['cole_bilingue'] == 'S']
non_bilingual = df[df['cole_bilingue'] == 'N']
```

```python
palette = sns.color_palette("Set2")

# Box plots comparativos
plt.figure(figsize=(15, 8))
for i, col in enumerate(['punt_matematicas', 'punt_sociales_ciudadanas', 'punt_c_naturales', 'punt_lectura_critica', 'punt_global']):
    plt.subplot(2, 3, i+1)
    sns.boxplot(x='cole_bilingue_desc', y=col, data=df, palette=palette)
    plt.title(f'Comparación de {col}')
    #plt.grid(True)
plt.tight_layout()
plt.show()

# Violin plots comparativos
plt.figure(figsize=(15, 8))
for i, col in enumerate(['punt_matematicas', 'punt_sociales_ciudadanas', 'punt_c_naturales', 'punt_lectura_critica', 'punt_global']):
    plt.subplot(2, 3, i+1)
    sns.violinplot(x='cole_bilingue_desc', y=col, data=df, palette=palette)
    plt.title(f'Comparación de {col}')
    #plt.grid(True)
plt.tight_layout()
plt.show()
```

Prueba de diferencia de medias

```python
means = df.groupby('cole_bilingue_desc')[['punt_matematicas', 'punt_sociales_ciudadanas', 'punt_c_naturales', 'punt_lectura_critica', 'punt_global']].mean()
means.plot(kind='bar', figsize=(10, 6), color=palette)
plt.title('Comparación de medias de puntajes')
plt.ylabel('Puntaje promedio')
plt.xticks(rotation=0)
plt.legend(title='Puntajes')
plt.grid(True)
plt.show()
```

```python
means = df.groupby('cole_bilingue_desc')[['punt_matematicas', 'punt_sociales_ciudadanas', 'punt_c_naturales', 'punt_lectura_critica']].mean()
means.plot(kind='bar', figsize=(10, 6), color=palette)
plt.title('Comparación de medias de puntajes')
plt.ylabel('Puntaje promedio')
plt.xticks(rotation=0)
plt.ylim(30,70)
plt.legend(title='Puntajes')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import ttest_ind
####--------------
#Prueba t para las diferencias de medias
# HO: u1 = u2 --> u1 - u2 = 0
# H1: u1 != u2 --> u1 - u2 != 0
####--------------
for col in ['punt_matematicas', 'punt_sociales_ciudadanas', 'punt_c_naturales', 'punt_lectura_critica', 'punt_global']:
    t_stat, p_val = ttest_ind(bilingual[col], non_bilingual[col], equal_var=False)
    print(f'Test t para {col}: estadístico t = {t_stat:.2f}, p-value = {p_val:.4f}')
```
